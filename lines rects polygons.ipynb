```python
import folium
```

```python
map = folium.Map(location=[27.620759026376376, 85.53959291408165], zoom_start=10)

# add separate layers for rectangles, lines, and polygons
rectangle_layer = folium.FeatureGroup(name="Rectangle Layer").add_to(map)
line_layer = folium.FeatureGroup(name="Line Layer").add_to(map)
polygon_layer = folium.FeatureGroup(name="Polygon Layer").add_to(map)
# add a rectangle to the rectangle layer
folium.Rectangle([[27.658844605753657, 85.32427081887934], [27.620739269895495, 85.53959482337858]],
                 color='red',
                 fill=True, fill_color='blue', fill_opacity=0.6).add_to(rectangle_layer)

# add lines to the line layer
folium.PolyLine([(27.658844605753657, 85.324270818879340), 
                 (27.620739269895495, 85.53959482337858),
                 (27.70614456268394, 85.3148035457928),
                ],
                 color='red',
                 weight=4.0).add_to(line_layer)

# add a polygon the the polygon layer
folium.Polygon([(27.73534209833566, 85.28080277528512), 
                 (27.620739269895495, 85.53959482337858),
                 (27.76328206411739, 85.44620962594186),
                (27.641685694197413, 85.36917658666898),
                ],
                 color='green', weight=4.0,
              fill=True, fill_color='orange', fill_opacity=0.4).add_to(polygon_layer)

# add layer control to the map
folium.LayerControl().add_to(map)
```

```python
map.save("lines rects polygons.html")
map
```
