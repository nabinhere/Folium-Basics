```python
import folium

# Create a map centered on a location
mymap = folium.Map(location=[51.5074, -0.1278], zoom_start=15)

# Define coordinates for the line
coordinates = [[51.5074, -0.1278], [51.5113, -0.1175], [51.5136, -0.1124]]

# Add the line to the map
folium.PolyLine(locations=coordinates, color='blue').add_to(mymap)

# Define the URL of the custom triangle image (replace 'triangle.png' with your image file)
icon_url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/0/05/Green_triangle.svg/220px-Green_triangle.svg.png'

# Add a marker with the custom triangle icon at a specific location
folium.Marker(location=[51.5074, -0.1278], icon=folium.CustomIcon(icon_url, icon_size=(20, 20))).add_to(mymap)
# Save the map
mymap.save('line_map.html')
```
