```python
import folium
```

```python
# create a map object centered around the given location
map = folium.Map(location = (27.620748965935363, 85.53959334996243), zoom_start=10)

# create a layer to add shapes(circles in this example)
# This layer can be turned off and on later
shapes_layer = folium.FeatureGroup(name="shapes").add_to(map)

# Latitudes, longitudes, and radius data of circles to be plotted
circle_data = [
    [27.620748965935363, 85.53959334996243, 400],
    [27.706064646304604, 85.31480635706637, 500],
    [27.661979049123666, 85.3255486653833, 600]
]

# add all circles to shapes layer
for circle in circle_data:
    folium.Circle(location = [circle[0], circle[1]],
                  radius=circle[2],
              color = 'blue',
              fill = True, fill_opacity = 1.0, fill_color = 'red',
             tooltip = 'Low voltage bus no.5',
             popup=folium.Popup("""<h2>This is a popup</h2><br/>
              This is a <b>new line</b><br/>
              <img src="https://www.w3schools.com/html/pic_trulli.jpg" alt="Trulli" style="max-width:100%;max-height:100%">""", max_width=1000)
                ).add_to(shapes_layer)

popup_text = 'hello hello hello hello hello hello hello hello hello hello hello hello'
folium.Circle(location = [27.672317019418063, 85.42827363785975],
                  radius=400,
              color = 'blue',
              fill = True, fill_opacity = 1.0, fill_color = 'red',
             tooltip = 'Low voltage bus no.5',
             popup=folium.Popup(popup_text, max_width=1000)
                ).add_to(shapes_layer)

# add layer control so that the layers can be turned on and off
folium.LayerControl().add_to(map)
```
